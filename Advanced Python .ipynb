# Python Inheritance

```python
class vehicle:
    def __init__(self,NoOfWheels, color, fuel, speed):
        self.NoOfWheels = NoOfWheels
        self.color = color
        self.fuel = fuel
        self.speed = speed
    
    def start(self):
        print(self.speed+", "+"Wizzzzzzzzzz")

    def brake(self):
        print("STOP!")

pride_saipa = vehicle(4, "red", "gasoline", "10km/hr")
#pride_saipa
#pride_saipa.speed
pride_saipa.start()
```

# Create a Child

```python
class car(vehicle):
    pass


peugeot405 = car(4, "black", "gasoline", "70km/hr")
peugeot405.start()
```

When you add the __init__() function, the child class will no longer inherit the parent's __init__() function: overriding! but still contain the methodes!

Check it out below:

```python
class car(vehicle):
    def __init__(self,NoOfDoors):
        self.NoOfDoors = NoOfDoors

```

```python
#peugeot206 = car(4, "black", "gasoline", "70km/hr")
peugeot206 = car(4)

peugeot206
peugeot206.brake()
```

# True Inheritance!

```python
class car(vehicle):
    def __init__(self,NoOfWheels, color, fuel, speed, NoOfDoors):
        super().__init__(NoOfWheels, color, fuel, speed) 
        self.NoOfDoors = NoOfDoors
```

```python
peugeot2008 = car(4, "black", "gasoline", "70km/hr", 2)
```

```python
peugeot2008
```
